**Portfolio task 1-R Portfolio (Data transformation)**

```r
library("tidyverse") #adding library 
```

```r
MyData<-tidyr::who#storing data from tidyr::who to MyData
```

```r
who1<-pivot_longer(MyData,c('new_sp_m014':'newrel_f65'), names_to="key", values_to="cases", values_drop_na = TRUE)#collecting all the columns from _new_spm014 to _newrelf65
```

```r
who1
```

```r
who2 <-who1 %>%mutate(key = stringr::str_replace(who1$key, 'newrel', 'new_rel'))#replacing newrel to new_rel
who2
```

```r
who3<-who2%>%separate(key, c("new", "type", "sexage"), sep = "_")
#%>% allow us to pass result of one who2 into function separate(key, c("new", "type", "sexage"), sep = "_") 
who3
```

```r
who4 <-separate(who3,sexage, c('sex', 'age'), sep='_')# seprating sex and age from the dataset who3 and storing to dataset who4
who4
```

```r
write.csv(who4, "who4.csv")#saving as who4.csv
```

```r
#Volkswagen brand offer best png in city with 35
ggplot(data = mpg)+
    geom_point(mapping = aes(x = manufacturer, y = cty,col="yellow"))
```

```r
#volkswagen offer best mpg in highway with mpg approximately equal to 44

ggplot(data = mpg)+
    geom_point(mapping = aes(x =manufacturer, y = hwy,col="red"))
```

```r
#Pickup type of car offers lower mpg in city 
ggplot(data = mpg)+
    geom_point(mapping = aes(x = displ, y = cty),col="violet")+
    facet_wrap(~ class, nrow = 4, ncol = 4)
```

```r
# car with displ (4,f) give best mpg performance in highway  as its performance rate is more than 40
ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = hwy,col="blue")) +
    facet_grid(drv ~ cyl)
```

```r
# car (4,f)  give best mpg performance in city  as its performance rate is approx 35

ggplot(data = mpg) +
    geom_point(mapping = aes(x = displ, y = cty),col="red") +
    facet_grid(drv ~ cyl)
```
